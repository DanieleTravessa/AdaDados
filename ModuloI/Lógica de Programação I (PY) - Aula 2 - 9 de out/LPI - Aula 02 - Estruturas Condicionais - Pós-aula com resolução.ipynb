## Estrutura condicional

Em nossos códigos, usualmente queremos que, dependendo dos valores das variáveis, a gente tenha respostas diferentes.
Em Python, faremos isso com o auxílio de três nomes especiais: `if`, `else` e `elif`(else if).

Estes comandos funcionam da seguinte forma:

```python
nota = 6
presenca = 0.60
# Se a pessoa tirou mais que 7, está aprovada.  
if nota > 7:
    print('Aprovado')
# Caso contrário, se a nota for maior que 5 E a presença maior que 75%, está aprovada.
elif nota >= 5 and presenca >= 0.75:
    print('Aprovado')
# Caso contrário, está reprovada.
else: 
    print('Reprovado')
```

```python
# REGRAS DO SISTEMA
# Se a pessoa tirou mais de 7, está aprovada.
# Se a pessoa teve menos que 75% de presença, está reprovada.
# Se a pessoa teve menos de 5, está reprovada.
# Caso contrário, está de recuperação. (nota entre 5 e 7 e presença maior que 75%)

if presenca < 0.75 or nota < 5:
    print('Reprovado')
elif nota > 7:
    print('Aprovado')
else:
    print('Recuperação')
```

Perceba que, quando usamos o `if`, precisamos nos atentar em algumas novidades:
- Precisamos definir uma condição. Essa condição deve retornar um valor booleano. Isto é, dizer se a condição é verdadeira ou não. Caso a condição seja verdadeira, o código realiza o que está dentro deste bloco. Caso contrário, ele apenas "ignora" o que está dentro.

- Precisamos usar o `:`. 

- Precisamos definir o que acontece se a condição é verdadeira.

```python
v = None 

# Se v não é vazio, então escrevo Eba.
if v:
    print('Eba')
```

8. Um produto vai sofrer aumento de acordo com a Tabela 1 abaixo. Faça um programa que peça para o usuário digitar o valor do produto de acordo 
com o preço antigo e escreva uma das mensagens da Tabela 2, de acordo com o preço reajustado:

Tabela 1

| Preço Antigo         | % de aumento |
|----------------------|--------------|
| Até 50 reais         | 5%           |
| Entre 50 e 100 reais | 10%          |
| De 100 a 150 reais   | 13%          |
| Acima de 150 reais   | 15%          |

Tabela 2

| Preço Novo            | Mensagem   |
|-----------------------|------------|
| Até 80 reais          | Barato     |
| Entre 80 e 115 reais  | Razoável   |
| Entre 115 e 150 reais | Normal     |
| Entre 150 e 170 reais | Caro       |
| Acima de 170 reais    | Muito Caro |

```python
# Solicita ao usuário que digite o valor do produto de acordo com o preço antigo
valor_antigo = float(input("Digite o valor do produto (em reais) de acordo com o preço antigo: "))

# Determina o percentual de aumento com base na Tabela 1
if valor_antigo <= 50:
    percentual_aumento = 5
elif valor_antigo <= 100:
    percentual_aumento = 10
elif valor_antigo <= 150:
    percentual_aumento = 13
else:
    percentual_aumento = 15

# Calcula o preço reajustado
preco_novo = valor_antigo * (1 + percentual_aumento / 100)

# Determina a mensagem com base na Tabela 2
if preco_novo <= 80:
    mensagem = "Barato"
elif preco_novo <= 115:
    mensagem = "Razoável"
elif preco_novo <= 150:
    mensagem = "Normal"
elif preco_novo <= 170:
    mensagem = "Caro"
else:
    mensagem = "Muito Caro"

# Exibe a mensagem correspondente ao preço reajustado
print(f"O preço reajustado é de R${preco_novo:.2f} e a classificação é: {mensagem}")
```

* Burguer light: pão, hamburguer, alface, tomate.

```python
# Estoques de ingredientes (representados como listas)
estoque_pao = ["pão"] * 10
estoque_hamburguer = ["hamburguer"] * 5
estoque_queijo = ["queijo"] * 8
estoque_alface = ["alface"] * 12
estoque_tomate = ["tomate"] * 7
estoque_bacon = ["bacon"] * 3

# Cardápio (representado como uma lista de listas)
cardapio = [
    ["Chessburguer", ["pão", "hamburguer", "queijo"]],
    ["Burguer completo", ["pão", "hamburguer", "queijo", "alface", "tomate", "bacon"]],
    ["Burguer light", ["pão", "hamburguer", "alface", "tomate"]],
]

# Pedido do cliente
pedido = "Burguer completo"

# Verificar a disponibilidade de ingredientes
ingredientes_pedido = []
falta_ingredientes = []

# Encontrar os ingredientes do pedido no cardápio
for item in cardapio:
    if item[0] == pedido:
        ingredientes_pedido = item[1]
        break

# Verificar a disponibilidade de cada ingrediente
for ingrediente in ingredientes_pedido:
    if ingrediente == "pão" and estoque_pao:
        estoque_pao.pop()
    elif ingrediente == "hamburguer" and estoque_hamburguer:
        estoque_hamburguer.pop()
    elif ingrediente == "queijo" and estoque_queijo:
        estoque_queijo.pop()
    elif ingrediente == "alface" and estoque_alface:
        estoque_alface.pop()
    elif ingrediente == "tomate" and estoque_tomate:
        estoque_tomate.pop()
    elif ingrediente == "bacon" and estoque_bacon:
        estoque_bacon.pop()
    else:
        falta_ingredientes.append(ingrediente)

# Verificar se o pedido pode ser atendido
if not falta_ingredientes:
    print(f"Pedido de {pedido} aceito. Os ingredientes estão sendo preparados.")
else:
    print(f"Desculpe, não podemos atender ao pedido de {pedido} devido à falta de ingredientes: {', '.join(falta_ingredientes)}.")
```
