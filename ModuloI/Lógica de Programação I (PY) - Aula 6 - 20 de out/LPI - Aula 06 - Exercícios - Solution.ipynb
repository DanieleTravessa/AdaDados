2. Suponha que você esteja desenvolvendo um sistema para cadastro de adoção de animais. Escreva um código que receba como parâmetros o nome do animal, o seu peso e a idade. Realize o controle para que o nome seja sempre uma palavra (sem números), o peso seja sempre um valor real e a idade sempre um valor inteiro e faça um print para indicar quando houver algum erro. 

````python
def cadastrar_animal(nome, peso, idade):
    if not nome.isalpha():
        print("Erro: O nome deve conter apenas letras.")
        return

    try:
        peso = float(peso)
    except ValueError:
        print("Erro: O peso deve ser um valor real (número com ponto ou vírgula).")
        return

    try:
        idade = int(idade)
    except ValueError:
        print("Erro: A idade deve ser um valor inteiro.")
        return

    # Se todas as validações passaram, os dados são válidos
    print("Cadastro bem-sucedido:")
    print("Nome: ", nome)
    print("Peso: ", peso)
    print("Idade: ", idade)

# Exemplo de uso:
nome = input("Nome do animal: ")
peso = input("Peso do animal: ")
idade = input("Idade do animal: ")

cadastrar_animal(nome, peso, idade)
````

3. No mercado financeiro, é comum verificar as variações de um ticker ao longo do dia. Crie um código que mostre a quantidade de vezes que esse ticker aumentou relativamente o seu valor. 

Considere que os últimos valores cotados deste papel estão contidos em uma lista com 24 valores. 

````python
# Lista com os valores cotados do ticker
valores_cotados = [100, 105, 110, 108, 112, 115, 113, 118, 120, 122, 121, 124, 126, 125, 128, 132, 130, 135, 137, 138, 136, 139, 141, 140]

# Inicialização do contador de aumentos relativos
aumentos_relativos = 0

# Iterar sobre os valores e contar os aumentos relativos
for i in range(1, len(valores_cotados)):
    if valores_cotados[i] > valores_cotados[i - 1]:
        aumentos_relativos += 1

# Imprimir o resultado
print(f"O ticker aumentou relativamente {aumentos_relativos} vezes.")
````

4. Uma empresa de confecção de vergalhões está automatizando a esteira de produção e gostaria de aumentar o controle de qualidade de seus produtos. Para isto, devemos criar um sistema que avalie se o comprimento do vergalhão está dentro do padrão estabelecido pela empresa, que varia entre 1 e 4 metros. Produtos que não passam por este teste de qualidade são descartados e considerados reprovados. 

Crie uma código que gere duas listas, uma contendo apenas os produtos que passaram pelo teste de qualidade e outra contendo os produtos reprovados. No final, mostre na tela:

- o percentual de produtos que passaram no teste.

- o comprimento **mediano** dos produtos aprovados.

- o comprimento máximo dos produtos reprovados.

- a diferença entre o maior e o menor produto.

Considere que existe uma lista `prods` contendo os comprimentos dos vergalhões, em metros. 

````python
# Lista de comprimentos dos vergalhões
prods = [3.5, 2.8, 4.2, 1.9, 3.0, 4.5, 0.8, 3.3, 2.1, 4.8, 1.5]

# Inicialização das listas de produtos aprovados e reprovados
aprovados = []
reprovados = []

# Avaliar cada vergalhão
for comprimento in prods:
    if 1.0 <= comprimento <= 4.0:
        aprovados.append(comprimento)
    else:
        reprovados.append(comprimento)

# Calcular o percentual de produtos aprovados
percentual_aprovados = (len(aprovados) / len(prods)) * 100

# Calcular o comprimento mediano dos produtos aprovados
aprovados.sort()
comprimento_mediano = aprovados[len(aprovados) // 2]

# Calcular o comprimento máximo dos produtos reprovados
comprimento_maximo_reprovados = max(reprovados)

# Calcular a diferença entre o maior e o menor produto
diferenca_max_min = max(prods) - min(prods)

# Imprimir os resultados
print("Produtos Aprovados:", aprovados)
print("Produtos Reprovados:", reprovados)
print(f"Percentual de Produtos Aprovados: {percentual_aprovados:.2f}%")
print(f"Comprimento Mediano dos Produtos Aprovados: {comprimento_mediano} metros")
print(f"Comprimento Máximo dos Produtos Reprovados: {comprimento_maximo_reprovados} metros")
print(f"Diferença entre o Maior e o Menor Produto: {diferenca_max_min} metros")
````

5. Sabemos que uma lista pode conter variáveis de diferentes tipos. Crie um programa que divida uma lista contendo variáveis do tipo `int`, `float` e `str` em três listas distintas, cada um contendo um tipo de variável. O programa deve mostrar na tela as três listas geradas.

Exemplo:


````python
# Lista contendo variáveis de diferentes tipos
lista_original = [1, 2.5, 'três', 4, 'cinco', 6.0, 'sete', 8]

# Listas para cada tipo de variável
lista_int = []
lista_float = []
lista_str = []

# Iterar pela lista original
for elemento in lista_original:
    if isinstance(elemento, int):
        lista_int.append(elemento)
    elif isinstance(elemento, float):
        lista_float.append(elemento)
    elif isinstance(elemento, str):
        lista_str.append(elemento)

# Imprimir as três listas geradas
print("Lista de inteiros:", lista_int)
print("Lista de floats:", lista_float)
print("Lista de strings:", lista_str)
````

6. Vamos fazer um jogo de adivinhação. Este jogo é feito para duas pessoas, e é jogado em duas etapas. Na primeira etapa, o primeiro jogador deve fornecer um número inteiro entre 0 e 500. O segundo jogador, sem saber este número, deve adivinhar até conseguir acertar, sendo contabilizado o número de tentativas até acertar o número. A cada adivinhação, o programa deve retornar se o número **escolhido** é maior ou menor do que o fornecido pelo primeiro jogador. 

Na segunda etapa, o jogo é invertido (i.e. o segundo jogador fornece o número inteiro e o primeiro jogador deve adivinhá-lo). O vencedor deste jogo será a pessoa que conseguir, no final, adivinhar o número com o menor número de rodadas.

Crie um código que reproduza o jogo, mostrando na tela o vencedor da partida. 

````python
# Função para adivinhar o número
def adivinhar_numero(numero_secreto, limite_inferior, limite_superior):
    tentativas = 0

    while True:
        tentativa = int(input(f"Tente adivinhar o número entre {limite_inferior} e {limite_superior}: "))
        tentativas += 1
        if tentativa < numero_secreto:
            print("O número é maior. Tente novamente.")
        elif tentativa > numero_secreto:
            print("O número é menor. Tente novamente.")
        elif tentativa < limite_inferior or tentativa > limite_superior:
            print("O número está fora do intervalo. Tente novamente.")
        else:
            print(f"Parabéns! Você acertou o número {numero_secreto} em {tentativas} tentativas.")
            return tentativas

# Primeira etapa: O primeiro jogador fornece o número
print("Primeira Etapa - Jogador 1 fornece o número")
numero_jogador1 = int(input("Jogador 1, forneça um número entre 0 e 500: "))

# Segunda etapa: O segundo jogador adivinha o número do primeiro jogador
print("\nSegunda Etapa - Jogador 2 adivinha o número de Jogador 1")
tentativas_jogador2 = adivinhar_numero(numero_jogador1, 0, 500)

# Terceira etapa: O segundo jogador fornece o número
print("\nTerceira Etapa - Jogador 2 fornece o número")
numero_jogador2 = int(input("Jogador 2, forneça um número entre 0 e 500: "))

# Quarta etapa: O primeiro jogador adivinha o número do segundo jogador
print("\nQuarta Etapa - Jogador 1 adivinha o número de Jogador 2")
tentativas_jogador1 = adivinhar_numero(numero_jogador2, 0, 500)

# Determinar o vencedor
if tentativas_jogador1 < tentativas_jogador2:
    print("Jogador 1 é o vencedor! Ele adivinhou o número com menos tentativas.")
elif tentativas_jogador2 < tentativas_jogador1:
    print("Jogador 2 é o vencedor! Ele adivinhou o número com menos tentativas.")
else:
    print("É um empate! Ambos os jogadores adivinharam o número em um número igual de tentativas.")
````

7. Um outro jogo que podemos implementar em Python é o jogo do Soma 100. Neste jogo, três pessoas podem participar. A cada rodada, cada um dos jogadores deve escolher, secretamente (neste caso, usando o `input` mas supondo que os outros jogadores não conseguem ver o núermo) um número entre 1 e 10 para ser adicionado ao montante conjunto `n`, que inicialmente tem o valor 0. O jogo termina quando, na rodada, o usuário ultrapassar 100, e o vencedor será aquele que foi o responsável pelo término do jogo. 

Crie um código que reproduza o jogo e mostre na tela o vencedor da partida.

````python
# Função para a escolha do número secreto por cada jogador
def escolher_numero(jogador):
    while True:
        try:
            numero = int(input(f"Jogador {jogador}, escolha um número entre 1 e 10: "))
            if 1 <= numero <= 10:
                return numero
            else:
                print("Por favor, escolha um número entre 1 e 10.")
        except ValueError:
            print("Por favor, insira um número válido entre 1 e 10.")

# Inicialização das variáveis
n = 0
rodada = 1
vencedor = None
pontos_jogador = [0, 0, 0]
while n < 100:
    print(f"\nRodada {rodada}\n")
    
    for jogador in range(0, 2):
        numero = escolher_numero(jogador)
        pontos_jogador[jogador] += numero
        print(f"Jogador {jogador + 1} escolheu {numero}. Montante atual: {pontos_jogador[jogador]}")
        
    
    if max(pontos_jogador) >= 100:
        # Checa se há mais de um vencedor
        if pontos_jogador.count(max(pontos_jogador)) > 1:
            # Recupera os indexes onde isso ocorre 
            vencedor = [i + 1 for i, x in enumerate(pontos_jogador) if x == max(pontos_jogador)]
        else:
            vencedor = [pontos_jogador.index(max(pontos_jogador)) + 1]
        break

    rodada += 1

if len(vencedor) == 1:
    print(f"\nJogador {vencedor[0]} venceu! Ele/ela ultrapassou 100 e encerrou o jogo.")
else:
    print(f"\nÉ um empate!. Os jogadores {', '.join(vencedor)} ultrapassaram 100 e encerraram o jogo.")
````
